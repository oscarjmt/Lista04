```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_gaussians_sum(k, sigma, seed=False):
    if seed:
        np.random.seed(seed)
    
    x_vals = np.random.uniform(0, 8, size=(k, 2))
    
    def f(x):
        f_sum = 0
        for i in range(k):
            f_sum += np.exp(- (np.linalg.norm(x - x_vals[i], ord=2) ** 2) / (2 * sigma))
        return -f_sum
    
    return f
```

```python
def plot_function_2d(f, x_range=(-1,9), y_range=(-1, 9), num_points=100, levels=10, figsize=(8, 6), show=True):
    x = np.linspace(x_range[0], x_range[1], num_points)
    y = np.linspace(x_range[0], x_range[1], num_points)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    for i in range(len(x)):
        for j in range(len(y)):
            Z[i, j] = f(np.array([X[i, j], Y[i, j]]))

    plt.figure(figsize=(8, 6))
    plt.contour(X, Y, Z, levels=levels)
    plt.colorbar()
    
    if show:
        plt.show()
```

```python
f = create_gaussians_sum(8, 1)
plot_function_2d(f)
```
